<h3 dir=rtl>
4. نمودار scatter 3d plot به شکل تعاملی تولید کنید. یعنی پلات به شکلی باشد که
بتوان در زوایای مختلف آن را چرخاند.
</h3>

```python
from mpl_toolkits.mplot3d import Axes3D

%matplotlib widget

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
