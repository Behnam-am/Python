```python
plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=.75)

plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.title('Scatter PLot')
plt.xlabel('Log View Count')
plt.ylabel('Log Total Likes')

plt.show()
```
