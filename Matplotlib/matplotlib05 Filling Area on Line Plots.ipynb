```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color='black', linestyle='--', label='all devs')
plt.plot(ages, py_salaries, label='python')

plt.fill_between(ages, py_salaries, alpha=0.25)
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='black', linestyle='--', label='all devs')
plt.plot(ages, py_salaries, label='python')

overal_median = 57287
plt.fill_between(ages, py_salaries, overal_median, alpha=0.25)
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='black', linestyle='--', label='all devs')
plt.plot(ages, py_salaries, label='python')

overal_median = 57287
plt.fill_between(ages, py_salaries, overal_median, 
                 where=(py_salaries > overal_median),
                 interpolate=True, alpha=0.25)
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='black', linestyle='--', label='all devs')
plt.plot(ages, py_salaries, label='python')

overal_median = 57287
plt.fill_between(ages, py_salaries, overal_median, 
                 where=(py_salaries > overal_median),
                 interpolate=True, color='green', alpha=0.35)
plt.fill_between(ages, py_salaries, overal_median, 
                 where=(py_salaries <= overal_median),
                 interpolate=True, color='red', alpha=0.35)
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='black', linestyle='--', label='all devs')
plt.plot(ages, py_salaries, label='python')

overal_median = 57287
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries > dev_salaries),
                 interpolate=True, color='green', alpha=0.35, label='more than average')
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.35, label='less than average')
plt.legend()
plt.show()
```
